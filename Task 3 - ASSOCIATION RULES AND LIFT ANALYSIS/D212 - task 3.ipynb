```python
import pandas as pd
import numpy as np
import jedi
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Visual libraries
import matplotlib.pyplot as plt

#Show all Columns and Rows
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
# Load data set
df = pd.read_csv('teleco_market_basket.csv')
```

```python
df.shape
```

```python
df.head(10)
```

```python
#Drop empty rows
df = df.dropna(subset= ['Item01'])
```

```python
# Converting data to list of list
trans = []
for i in range(0, df.shape[0]):
    trans.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
# Transforming the list of list to numpy
TE= TransactionEncoder()
newlist = TE.fit(trans).transform(trans)

# Row are logical and columns are purchased items
newDataSet = pd.DataFrame(newlist, columns=TE.columns_)
```

```python
# View of the new data set
newDataSet.shape
```

```python
# View of the new data set
newDataSet.head()
```

```python
newDataSet.to_csv('D212_cleanDataSet.csv')
```

```python
# Inspecting columns
for i in newDataSet:
    print(i)
```

```python
# Remove nan column
newDataSet = newDataSet.drop(['nan'], axis=1)
```

### Analysis

```python
# Apply Appriori Analysis
appriori_rules = apriori(newDataSet, min_support=0.02, use_colnames=True)
appriori_rules.head()
```

```python
# Plot ks vs inertias
appriori_results = list(appriori_rules)
appriori_results
```

```python
# Association Rules
assoc_rule = association_rules(appriori_rules, metric='lift', min_threshold=1)
assoc_rule.head(10)
```

```python
# Pruning rule
prune_rule = assoc_rule[assoc_rule['confidence']>.02]
print(len(prune_rule))
```

```python
# Sorting Rules
sort_rules = prune_rule.sort_values('lift', ascending=False,).head()
sort_rules
```
